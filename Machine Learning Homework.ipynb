```python
import pandas as pd
```

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head(100)
```

```python
# Set features. This will also be used as your x values.
#selected_features = df[['names', 'of', 'selected', 'features', 'here']]
df.columns
```

```python
df["koi_disposition"].value_counts()
```

```python
cleanup_nums = {"koi_disposition": {"FALSE POSITIVE": 0, "CONFIRMED": 1, "CANDIDATE": 2}}
```

```python
df.replace(cleanup_nums, inplace=True)
df.head()
```

```python
df["koi_disposition"].value_counts()
```

```python
df.corr()
```

## There does not appear to be much in the way of significantly strong feature correlation between any two features in the dataset.

```python
#Quitamos todas las columnas que contienen error para el data frame como 'koi_slogg_err1'
exoplanet_df=df.filter(regex =('^((?!err).)*$'),axis=1)
```

```python
#Create a Train Test Split
y = exoplanet_df["koi_disposition"]
X = exoplanet_df.drop(columns=["koi_disposition"])
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
X_train.head()
```

```python
#Pre-processing
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from sklearn.svm import SVC 
model = SVC(gamma = "auto", kernel='linear')
model.fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5, 10, 50],
              'gamma': [0.0001, 0.0005, 0.001, 0.005]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
grid.fit(X_train_scaled, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
import joblib
filename = 'finalized_model.sav'
joblib.dump(model, filename)
```

```python
loaded_model = joblib.load(filename)
result = loaded_model.score(X_test_scaled, y_test)
print(result)
```

```python
from sklearn.metrics import classification_report
```

```python
grid_model = grid.predict(X_test_scaled)
```

```python
print(classification_report(y_test, grid_model))
```
